```python
df.cast.str.split(",", expand=True)
```

```python
actor_count = df.copy()
actor_count = pd.concat([actor_count, df.cast.str.split(",", expand=True)],axis=1)
actor_count = actor_count.melt(id_vars='title', value_vars=range(49), value_name='actor')
actor_count = actor_count[actor_count['actor'].notna()]
```

```python
actor_count = actor_count['actor'].value_counts()
actor_count = actor_count[actor_count > 15]
```

```python
actor_count.plot(kind='bar',figsize=(20,10))
```

```python
px.histogram(actor_count, x='actor')
```
